```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
ft = pd.read_csv('C:/for_python/istherecorrelation1.csv',sep=';')

# Adjust data frame
ft['NL Beer consumption [x1000 hectoliter]'] = ft['NL Beer consumption [x1000 hectoliter]'] * 1000
ft = ft.rename(columns ={'NL Beer consumption [x1000 hectoliter]':'NL Beer consumption [in hectoliters]'})

ft['WO [x1000]'] = ft['WO [x1000]'].str.replace(',','.')
ft['WO [x1000]'] = pd.to_numeric(ft['WO [x1000]'])

ft['WO [x1000]'] = ft['WO [x1000]'] * 1000
ft = ft.rename(columns ={'WO [x1000]':'WO students'})


#plotting number of WO students in the Neatherlands over the years 
plt.figure(dpi = 300)
plt.plot(ft.iloc[:,0],ft.iloc[:,1])
plt.grid()
plt.title('WO students over the years in the Neatherlands')

#potting beer consumption 
plt.figure(dpi = 300)
plt.plot(ft.iloc[:,0],ft.iloc[:,2])
plt.grid()
plt.title('Beer consumption over the years in the Neatherlands')

#plotting both plots on one - with two y-axes to account for different units

fig, ax1 = plt.subplots(dpi=300)


ax1.set_xlabel('Year')
ax1.set_ylabel('Alcohol consumption in the Netherlands (in hektoliters)', color='tab:blue')
ax1.plot(ft['Year'], ft['NL Beer consumption [in hectoliters]'], color='tab:blue', marker='o', label='Alcohol consumption in the Netherlands (hektoliters)')
ax1.tick_params(axis='y', labelcolor='tab:blue')


ax2 = ax1.twinx()
ax2.set_ylabel('Number of WO Students', color='tab:orange')
ax2.plot(ft['Year'], ft['WO students'], color='tab:orange', marker='s', label='Number of WO Students')
ax2.tick_params(axis='y', labelcolor='tab:orange')


plt.title('Alcohol Consumption (Netherlands) / Number of WO Students')
plt.grid()

plt.savefig('fig.png', dpi=300)
plt.show()
```
