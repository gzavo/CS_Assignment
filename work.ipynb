```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("istherecorrelation.csv", sep=';')

xaxis = data['Year']
beer = data['NL Beer consumption [x1000 hectoliter]']
students = data['WO [x1000]'].str.replace(',', '.').astype(float)
yaxis = beer / students

plt.figure(figsize=(10,6), dpi=300)  
plt.plot(xaxis, yaxis, linestyle='-', color='r')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Beer Consumption [x1 hectoliter] per WO Student', fontsize=12)
plt.title('Beer Consumption per Student Over the Years', fontsize=14)
plt.legend()

plt.savefig('plot.png', dpi=300)
plt.show()
```
