```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\Alien\Downloads\istherecorrelation.csv") 

df
```

```python
year_list = []
WO_list = []
Beer_list= []

for index, row in df.iterrows():
    year_list.append(float(index[0:4]))
    WO_list.append(float(index[5:8]) + (float(row[0][0]))/10)
    Beer_list.append(float(row[0][2:]))
    


fig, ax1 = plt.subplots()
fig.dpi = 300

ax1.set_xlabel('Year')
ax1.set_ylabel('WO students x1000')
ax1.set_ylim(0, 300)
ax1.plot(year_list, WO_list)

ax2 = ax1.twinx()
ax2.set_ylabel('Beer drunk x1000 Hectoliter')
ax2.plot(year_list, Beer_list, color='orange')
ax2.set_ylim(0, 12500)

fig.suptitle('Number of Students in higher education(WO) vs the ammount of Beer drunk Annually')

fig.show
fig.savefig('Assignment 1 graph', dpi=300)
# we already handled the x-label with ax1
```
