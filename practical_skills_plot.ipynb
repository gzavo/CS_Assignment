```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("istherecorrelation.csv",sep=";")
print(df)
```

```python
plt.figure(figsize=(8, 6))

#scatter plot with seaborn
sns.scatterplot(
    data=df,
    x="WO [x1000]",
    y="NL Beer consumption [x1000 hectoliter]",
    hue="Year",
    palette="viridis",
    s=100
)

# adding year labels next to each point
for i, row in df.iterrows():
    plt.text(
        row["WO [x1000]"] + 0.5,   #adding small horizontal offset to show the labels
        row["NL Beer consumption [x1000 hectoliter]"], 
        int(row["Year"]),
        fontsize=8
    )

#plotting titles
plt.title("Students in WO Education vs Beer Consumption (2006-2018)")
plt.xlabel("Students in WO Education [x1000]")
plt.ylabel("Beer Consumption [x1000 hl]")

#saving the figure picture
plt.savefig("scatter_plot_300dpi.png", dpi=300, bbox_inches='tight')
plt.show()
```
