```python
#Idea of SIR algorithm
import matplotlib as mlp
import matplotlib.pyplot as plt
import numpy as np
import csv
%matplotlib notebook

x_values = []
y_values = []
z_values = []
with open('istherecorrelation.csv', newline='') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=';', quotechar='|')
    spamreader.__next__()
    for row in spamreader:
        x_values.append(int(row[0].replace(',','.')))
        y_values.append(float(row[1].replace(',','.')))
        z_values.append(int(row[2].replace(',','.')))


plt.figure()
plt.plot(x_values,y_values,'g*')
plt.xlabel('year')
plt.ylabel('WO') 
plt.ylim(0,max(y_values))
plt.savefig('2d_1_plot.png', dpi=300)

print("Average WO:",np.mean(y_values),"STD WO:",np.std(y_values))

plt.figure()
plt.plot(x_values,z_values,'g*')
plt.xlabel('year')
plt.ylabel('hectoliter beer') 
plt.ylim(0,max(z_values))
plt.savefig('2d_2_plot.png', dpi=300)


print("Average Hectoliter Beer:",np.mean(z_values),"STD Hectoliter Beer:",np.std(z_values))
```
