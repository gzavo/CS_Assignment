```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('istherecorrelation.csv', sep = ';', decimal = ',')
```

```python
data.head()
```

```python
ax1 = data.plot(x = 'Year', y = ['WO [x1000]'], legend = False)
ax1.set_ylabel('WO [x1000]', color = 'C0')

ax2 = ax1.twinx()
ax2.spines['right'].set_position(('axes', 1.0))
data.plot(x = 'Year', y = 'NL Beer consumption [x1000 hectoliter]', ax=ax2, color = 'C1', legend = False)
ax2.set_ylabel('NL Beer consumption [x1000 hectoliter]', color = 'C1')

plt.savefig('istherecorrelation.png', bbox_inches='tight', dpi = 300)
```
