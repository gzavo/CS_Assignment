```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")
```

```python
# Read in the data
data = pd.read_csv('istherecorrelation.csv', delimiter=';', decimal=',')
```

```python
# plot data correlation matrix
sns.heatmap(data.corr(method='pearson'), cmap="YlGnBu", annot=True, )
plt.title('Correlation matrix')
plt.savefig('correlation_matrix.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(10, 5))
ax1.set_xlabel('Year')
ax1.set_ylabel('WO Students [x1000]', color='tab:blue')
ax1.tick_params(axis='y', colors='tab:blue')
ax2 = ax1.twinx()
ax2.set_ylabel('Beer consumption [x1000 hectoliter]', color='tab:red')
ax2.tick_params(axis='y', colors='tab:red')
sns.regplot(x="Year", y="WO [x1000]", data=data, ax=ax1, color='tab:blue')
sns.regplot(x="Year", y="NL Beer consumption [x1000 hectoliter]", data=data, ax=ax2, color='tab:red')
plt.title('WO student count vs Beer consumption in Netherlands')
plt.figtext(0.5, -0.05, '(incl. linear regression plots and 95% confidence intervals)', wrap=True, horizontalalignment='center', fontsize=10)
plt.savefig('wo_vs_beer.png', dpi=300, bbox_inches='tight')
plt.show()
```
