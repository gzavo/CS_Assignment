```python
import csv
import matplotlib.pyplot as plt
# Open and read the file
with open('istherecorrelation.csv', 'r') as file:
    data = file.readlines()

split_data = [line.strip().split(';') for line in data]

y, s, a = [], [], []
for row in split_data[1:]:
    y.append(int(row[0]))
    s.append(float(row[1].replace(',', '.')))
    a.append(int(row[2]))
```

```python
import matplotlib.pyplot as plt

fig, ax1 = plt.subplots()

ax1.set_xlabel('Year')
ax1.set_ylabel(str(split_data[0][1]), color='tab:blue')
ax1.plot(y, s, color='tab:blue', label=str(split_data[0][1]))
ax1.tick_params(axis='y')

ax2 = ax1.twinx()
ax2.set_ylabel(str(split_data[0][2]), color='tab:red')
ax2.plot(y, a, color='tab:red', label=str(split_data[0][2]))
ax2.tick_params(axis='y')

plt.subplots_adjust(top=0.9)
plt.savefig('wo_vs_beer_consumption.png', bbox_inches='tight', dpi=300)
plt.title('Amount of WO Students VS Beer Consumption NL')
plt.show()
```
