```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Read in the data with sep=';'
data = pd.read_csv('istherecorrelation.csv', sep=';') 
# replace commas with dots in the 'WO [x1000]' column
data['WO [x1000]'] = data['WO [x1000]'].str.replace(',', '.').astype(float)
```

```python
# Plot the data for correlation between WO and beer consumption with dpi=300
fig, ax = plt.subplots()

# Plot the first set of data
ax.plot(data['Year'], data['WO [x1000]'], label='WO [x1000]')
ax.set_xlabel('Year')
ax.set_ylabel('WO [x1000]')

# Create a twin axis to plot the second set of data
ax2 = ax.twinx()
ax2.plot(data['Year'], data['NL Beer consumption [x1000 hectoliter]'], 'r', label='NL Beer consumption [x1000 hectoliter]')
ax2.set_ylabel('NL Beer consumption [x1000 hectoliter]')

# Add the legend
fig.legend(loc='upper left')

# Save the figure with the desired resolution (e.g., 300 DPI)
plt.savefig('correlation.png', dpi=300)

# Display the plot
plt.show()
```
