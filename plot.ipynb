```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('istherecorrelation.csv', sep=';')

df.head()
```

```python
df.dtypes
```

```python
df['WO [x1000]'].replace(',','.', regex=True, inplace=True)
df['WO [x1000]'] = df['WO [x1000]'].astype(float)
df.dtypes
```

```python
plt.figure(figsize=(10,5), dpi=300)
plt.plot(df['Year'], df['WO [x1000]'].pct_change())
plt.plot(df['Year'], df['NL Beer consumption [x1000 hectoliter]'].pct_change())
plt.legend(['Waste of O2', 'NL beer consumption (hl)'])
plt.title('Waste of O2 vs. beer consumption')
plt.xlabel('Year')
plt.ylabel('x1000')
plt.show()
```
