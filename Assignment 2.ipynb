```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("istherecorrelation.csv", sep=";", decimal=',')
df_norm=df.copy()
for column in df_norm[['WO [x1000]', 'NL Beer consumption [x1000 hectoliter]']]:
    df_norm[column] = (df_norm[column] - df_norm[column].min()) / (df_norm[column].max() - df_norm[column].min())

fig=plt.figure(dpi=300)
df_norm.plot(x="Year", title = "Plot of WO and NL Beer consumption per year (normalised)", ax=plt.gca())

df.corr()
```
