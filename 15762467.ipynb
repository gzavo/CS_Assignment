```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = 'https://raw.githubusercontent.com/zoeazra/CS_Assignment/refs/heads/main/istherecorrelation.csv'
data = pd.read_csv(url, sep=';', decimal=',')

data.head()
```

```python
plt.plot(data['Year'], data['NL Beer consumption [x1000 hectoliter]'], marker='o', linestyle='-', color='b')
plt.xlabel('Year')
plt.ylabel('Beer consumption (x1000 hectoliter)')
plt.title('Plot of beer consumption in The Netherlands over a span of 12 years')

plt.savefig('beerplot.png', dpi=300)

plt.show()
```
