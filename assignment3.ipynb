```python
import matplotlib.pyplot as plt

# Data
years = [2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018]
wo = [205.9, 208.6, 212.7, 220.5, 233.2, 242.4, 245.4, 241.4, 250.2, 255.7, 261.2, 267.9, 280.1]
beer_consumption = [11402, 11492, 11450, 11502, 11474, 11480, 11452, 11484, 11555, 11601, 11731, 11862, 12048]

# Create a plot
fig, ax1 = plt.subplots()

# Plot WO on the left y-axis
ax1.set_xlabel('Year')
ax1.set_ylabel('WO [x1000]', color='tab:blue')
ax1.plot(years, wo, color='tab:blue', label='WO [x1000]')
ax1.tick_params(axis='y', labelcolor='tab:blue')

# Create a second y-axis for beer consumption
ax2 = ax1.twinx()
ax2.set_ylabel('NL Beer consumption [x1000 hectoliter]', color='tab:orange')
ax2.plot(years, beer_consumption, color='tab:orange', label='NL Beer consumption [x1000 hectoliter]')
ax2.tick_params(axis='y', labelcolor='tab:orange')

# Set title and show the plot
plt.title('WO vs NL Beer Consumption Over the Years')
plt.tight_layout()
plt.savefig('correlation_plot.png', dpi=300)
plt.show()
```
